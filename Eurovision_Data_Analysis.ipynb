```python
# Filter data for finals only
contestants_final = contestants[contestants['sf_num'].isnull()]
```

```python
# Group by year and country to sum total votes
votes_final = votes[votes['round'] == 'final']
```

```python
# Calculate the Herfindahl Index for each year
def calculate_hhi(year, df):
    total_votes = df['total_points'].sum()
    df['vote_share'] = df['total_points'] / total_votes
    hhi = (df['vote_share'] ** 2).sum()
    return hhi
```

```python
# Calculate the gap between winner and second-place finisher
def calculate_vote_gap(year, df):
    sorted_df = df.sort_values(by='total_points', ascending=False)
    if len(sorted_df) >= 2:
        gap = sorted_df.iloc[0]['total_points'] - sorted_df.iloc[1]['total_points']
    else:
        gap = None
    return gap
```

```python
# Initialize lists to store results
years = []
hhi_values = []
vote_gaps = []

for year in contestants_final['year'].unique():
    year_data = votes_final[votes_final['year'] == year]
    if not year_data.empty:
        hhi = calculate_hhi(year, year_data)
        gap = calculate_vote_gap(year, year_data)
        years.append(year)
        hhi_values.append(hhi)
        vote_gaps.append(gap)
```

```python
# Create DataFrame to store results
results = pd.DataFrame({
    'year': years,
    'hhi': hhi_values,
    'vote_gap': vote_gaps
})
```

```python
# Determine the year semi-finals were introduced
semi_final_intro_year = 2004
```

```python
# Calculate average HHI and vote gap before and after semi-final introduction
before_semi_final = results[results['year'] < semi_final_intro_year]
after_semi_final = results[results['year'] >= semi_final_intro_year]

average_hhi_before = before_semi_final['hhi'].mean()
average_hhi_after = after_semi_final['hhi'].mean()

average_gap_before = before_semi_final['vote_gap'].mean()
average_gap_after = after_semi_final['vote_gap'].mean()
```

```python
# Output the results
print("Average HHI before semi-finals:", average_hhi_before)
print("Average HHI after semi-finals:", average_hhi_after)
print("Average vote gap before semi-finals:", average_gap_before)
print("Average vote gap after semi-finals:", average_gap_after)

results
```

```python
# Filter for final odds only
betting_final = betting_offices[betting_offices['contest_round'] == 'final']
```

```python
# Convert odds to probabilities
betting_final['probability'] = 1 / betting_final['betting_score']
```

```python
# Aggregate probabilities using the harmonic mean
betting_final_agg = betting_final.groupby(['year', 'country_name']).agg({
    'probability': lambda x: len(x) / sum(1 / x)
}).reset_index()
```

```python
# Calculate HHI for each year
def calculate_hhi(df):
    return (df['probability'] ** 2).sum()
```

```python
hhi_by_year = betting_final_agg.groupby('year').apply(calculate_hhi).reset_index()
hhi_by_year.columns = ['year', 'hhi']
```

```python
# Calculate vote gap for each year
def calculate_vote_gap(df):
    sorted_probs = df['probability'].sort_values(ascending=False).values
    if len(sorted_probs) > 1:
        return sorted_probs[0] - sorted_probs[1]
    else:
        return None

vote_gap_by_year = betting_final_agg.groupby('year').apply(calculate_vote_gap).reset_index()
vote_gap_by_year.columns = ['year', 'vote_gap']
```

```python
# Merge HHI and vote gap data
competitiveness_metrics = pd.merge(hhi_by_year, vote_gap_by_year, on='year')

competitiveness_metrics.head(10)
```

```python
# Standardize country names (mapping to ensure names match)
country_mapping = {
    'at': 'Austria',
    'be': 'Belgium',
    'ch': 'Switzerland',
    'de': 'Germany',
    'dk': 'Denmark',
    'es': 'Spain',
    'fi': 'Finland',
    'fr': 'France',
    'gb': 'United Kingdom',
    'gr': 'Greece',
    'ie': 'Ireland',
    'il': 'Israel',
    'it': 'Italy',
    'lu': 'Luxembourg',
    'mt': 'Malta',
    'nl': 'Netherlands',
    'no': 'Norway',
    'pt': 'Portugal',
    'se': 'Sweden',
    'si': 'Slovenia',
    'ua': 'Ukraine'
}
```

```python

# Apply the mapping
vote_tallies['to_country'] = vote_tallies['to_country'].map(country_mapping)
```

```python
# Filter the data to include only the common years
common_years = set(aggregated_probabilities['year']).intersection(set(vote_tallies['year']))
aggregated_probabilities = aggregated_probabilities[aggregated_probabilities['year'].isin(common_years)]
vote_tallies = vote_tallies[vote_tallies['year'].isin(common_years)]
```

```python
# Merge aggregated probabilities with vote tallies
merged_data = pd.merge(aggregated_probabilities, vote_tallies, left_on=['year', 'country_name'], right_on=['year', 'to_country'])
```

```python
merged_data
```

```python
# Drop any rows with missing values
merged_data.dropna(subset=['probability', 'total_points'], inplace=True)
```

```python
# Calculate the Pearson correlation coefficient for each year
def safe_pearsonr(df):
    if len(df) < 2:
        return None
    return pearsonr(df['probability'], df['total_points'])[0]
```

```python
pearson_results = merged_data.groupby('year').apply(safe_pearsonr).reset_index(name='pearson_corr')
```

```python
pearson_results
```

```python
# Calculate the overall Pearson correlation
overall_pearson = pearsonr(merged_data['probability'], merged_data['total_points'])[0]
```

```python
# Calculate the rankings
merged_data['probability_rank'] = merged_data.groupby('year')['probability'].rank(ascending=False, method='min')
merged_data['vote_tally_rank'] = merged_data.groupby('year')['total_points'].rank(ascending=False, method='min')
```

```python
# Calculate the Spearman rank correlation for each year
def safe_spearmanr(df):
    if len(df) < 2:
        return None
    return spearmanr(df['probability_rank'], df['vote_tally_rank'])[0]
```

```python
spearman_results = merged_data.groupby('year').apply(safe_spearmanr).reset_index(name='spearman_corr')
```

```python
spearman_results
```

```python
# Calculate the overall Spearman correlation
overall_spearman = spearmanr(merged_data['probability_rank'], merged_data['vote_tally_rank'])[0]
```
